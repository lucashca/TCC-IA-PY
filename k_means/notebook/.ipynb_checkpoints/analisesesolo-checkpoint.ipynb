```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sb

%matplotlib inline
```

```python
df = pd.read_csv('../datasets/iris.csv')
print(df)
```

```python
sb.pairplot(df)
```

```python
X = np.array(df.drop('target',axis=1))
X
```

# K-Means

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3,random_state = 0,n_jobs=1)
```

```python
kmeans.fit(X)
```

```python
print(kmeans.labels_)
```

# Comparando as Classes Criadas

```python

df['K-classes'] = kmeans.labels_
```

```python
df
```

```python
sb.pairplot(df,hue='target')
```

```python
dfK =df.drop(columns=['target'])
dfK
```

```python
sb.pairplot(dfK,hue='K-classes')
```
